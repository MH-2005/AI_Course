```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

<h2 style="font-family: 'Vazir', Arial, sans-serif; color: white; text-align: center; margin-bottom: 20px; font-weight: bold; direction: rtl;"> درخواست دادن به url زیر و گرفتن محتوای درون آن </h2>

```python
url = "https://www.tripadvisor.com/TravelersChoice-Hotels"

chrome_options = Options()
chrome_options.add_argument("--start-maximized")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"
chrome_options.add_argument(f'user-agent={user_agent}')
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")

soup = None

try:
    service = ChromeService(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)

    print("Getting page...")
    driver.get(url)
    time.sleep(10)

    pageHtml = driver.page_source
    print("Page source retrieved. Parsing with BeautifulSoup...")

    soup = BeautifulSoup(pageHtml, "html.parser")
    print("Parsing complete.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    if 'driver' in locals():
        driver.quit()
        print("Driver closed.")

if soup == None:
    raise ValueError("No soup found.")
```

```python
# code for scraping item names
names = soup.find_all("h2", {"class": "biGQs _P fiohW BDkrC"})
names.insert(0, soup.find("h2" , "biGQs _P fiohW LSyRd"))
for element in names:
    print(element.text)
```

```python
# code for scraping hotel places
places = soup.find_all("a", {"class": "BMQDV _F Gv wSSLS SwZTJ FGwzt suezE"})[::2]
for element in places:
    print(element.text)
```

```python
# code for scraping hotel rates
rates = []

rate_span = soup.find('span', class_="IvARn")
if rate_span:
    rates.append(rate_span)

bubble_rates = soup.find_all('div', {'data-automation': 'bubbleRatingValue'})
rates.extend(bubble_rates)

for element in rates:
    print(element.text)
```

```python
# Code for scraping hotel links
links = []
for name in names:
    link = "https://www.tripadvisor.com" + name.parent['href']
    links.append(link)

for element in links:
    print(element)
```

<h2 style="font-family: 'Vazir', Arial, sans-serif; color: white; text-align: center; margin-bottom: 20px; font-weight: bold; direction: rtl;">ذخیره کردن اطلاعات اسکرپ شده</h2>

```python
# code for saving data that scraped
list_hotels = []
for i in range(len(names)):
    list_hotels.append([names[i].text, places[i].text, rates[i].text,links[i] ])
with open('Hotels.txt', 'w') as f:
    for item in list_hotels:
        f.write(str(item) + '\n')
```
