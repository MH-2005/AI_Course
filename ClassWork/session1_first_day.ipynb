<h1>"Modifying the tensor leads to changes in the image. I aim to adjust the tensor to achieve specific changes in the picture."</h1>

```python
image = cv.imread(r"Bird Speciees Dataset/BARN OWL/001.jpg")
img_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# کپی گرفتن از تصویر برای تغییر
tensor = img_rgb.copy().astype(np.float32)

# ----------------------------
# تغییرات: افزایش روشنایی
# ----------------------------
bright_tensor = tensor + 50  # افزایش مقدار پیکسل‌ها
bright_tensor = np.clip(bright_tensor, 0, 255).astype(np.uint8)  # محدود به بازه مجاز

# ----------------------------
# تغییرات دیگر (مثلاً معکوس رنگ):
# inverted_tensor = 255 - tensor
# ----------------------------

# نمایش تصویر اصلی و تغییر یافته
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Brightness Increased")
plt.imshow(bright_tensor)
plt.axis('off')

plt.tight_layout()
plt.show()
```

<p>
در اینجا می توانیم از HGV استفاده کنیم که برای افزایش سطح روشنایی مثل اینجا دیگر مشکل در مورد به هم خوردن رنگ ها نداشته باشیم 
</p>

```python
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# استخراج کانال‌های رنگی
Red_format = image_rgb[:, :, 0]
Green_format = image_rgb[:, :, 1]
Blue_format = image_rgb[:, :, 2]

# نمایش کانال‌ها
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(Red_format, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(Green_format, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(Blue_format, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```
