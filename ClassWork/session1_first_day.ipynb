```python
def sh (im):
    cv.imshow("image", im)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

<h1>Resize</h>

```python
OWL_IMAGE = cv.resize(image,(1000,1000),interpolation=cv.INTER_LINEAR)
```

```python
cv.imshow("image", OWL_IMAGE)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
for _ in range(12):
    OWL_IMAGE = cv.pyrUp(image)
```

```python
cv.imshow("image", OWL_IMAGE)
cv.waitKey(0)
cv.destroyAllWindows()
```

<h1>CROPPED</h1>

```python
cropped = image[100:200,0:100]
```

```python
cv.imshow("image", cropped)
cv.waitKey(0)
cv.destroyAllWindows()
```

<h1>Rotate</h1>

```python
rotated = cv.rotate(image,cv.ROTATE_90_CLOCKWISE)
```

```python
sh(rotated)
```

<h1>Flip</h1>

```python
flipped = cv.flip(image,-1)
```

```python
sh(flipped)
```
