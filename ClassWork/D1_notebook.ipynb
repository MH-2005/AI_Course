## Ectracting image information 

```python
# image_info = image.shape

def image_information(image):
    image_info = image.shape

    info = f"Height: {image_info[0]}, "
    info += f"width: {image_info[1]}, "
    
    if len(image_info) == 3:
        info += f"Channel: {image_info[2]}"
    else :
        info += f"channel: 0"

    return info
```

## Defining a function for further use

```python
def imshow(title= '', image= None, size= 10):
    W, H= image.shape[0], image.shape[1]
    aspect_ratio = H/W
    plt.figure(figsize= (aspect_ratio*size, size))
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
    
```

### Croping an image 
so far we know in opencv we map the pixel value to a matrix so we can make a cropped image from the main image . 

```python
bird_img = cv.imread("../Resources/bird.jpeg", cv.IMREAD_COLOR_RGB)
print(bird_img.shape)

cropped_image = bird_img[200:800, 400:1000]


plt.figure(figsize=[20, 5])
plt.subplot(121); plt.imshow(bird_img); plt.title("Original image")
plt.subplot(122); plt.imshow(cropped_image); plt.title("Cropped image")
plt.show()
```

## Rotating and flipping images

```python
bird_image = cv.cvtColor(bird_image,cv.COLOR_BGR2RGB)
rotated_image90 = cv.rotate(bird_image, cv.ROTATE_90_CLOCKWISE)
rotated_image180 = cv.rotate(bird_image, cv.ROTATE_180)
rotated_image270 = cv.rotate(bird_image, cv.ROTATE_90_COUNTERCLOCKWISE)

plt.subplot(131); plt.imshow(rotated_image90); plt.title('90 degree image')
plt.subplot(132); plt.imshow(rotated_image180); plt.title('180 degree image')
plt.subplot(133); plt.imshow(rotated_image270); plt.title('270 degree image')
```

```python
#x flip
#y flip
#x-y rotate
flipped_image = cv.flip(bird_image, -1)
imshow('flipped_image', flipped_image, 10)
```

```python
gray_scaled = cv.cvtColor(bird_image, cv.COLOR_RGB2GRAY)

plt.figure(figsize=[10,5])
plt.subplot(131);plt.imshow(gray_scaled, cmap='jet')
plt.subplot(132);plt.imshow(gray_scaled, cmap='hot')
plt.subplot(133);plt.imshow(gray_scaled, cmap='BuPu')

```

```python
inversed_image = 255 - gray_scaled
plt.figure(figsize=[10,5])
plt.subplot(131);plt.imshow(inversed_image, cmap='jet')
plt.subplot(132);plt.imshow(inversed_image, cmap='gray')
plt.subplot(133);plt.imshow(inversed_image, cmap='BuPu')
```
