```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import Image, display
def sh(img):
    display(img)
```

```python
def show_images(images, titles=None, cols=2, cmap=None, scale=4):
    """
    نمایش لیستی از تصاویر در یک گرید.

    Parameters:
    - images: لیستی از تصاویر (NumPy arrays)
    - titles: لیستی از عنوان‌ها (اختیاری)
    - cols: تعداد ستون‌ها (پیش‌فرض: 2)
    - cmap: اگر همه تصاویر خاکستری‌اند، می‌توان cmap='gray' استفاده کرد
    - scale: اندازه هر تصویر (برای بزرگ‌نمایی)
    """
    n_images = len(images)
    rows = (n_images + cols - 1) // cols
    plt.figure(figsize=(scale * cols, scale * rows))

    for i in range(n_images):
        plt.subplot(rows, cols, i + 1)

        img = images[i]

        # اگر تصویر رنگی BGR باشد، تبدیل به RGB
        if img.ndim == 3 and img.shape[2] == 3:
            img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

        if cmap:
            plt.imshow(img, cmap=cmap)
        else:
            # به صورت خودکار انتخاب شود
            if img.ndim == 2:
                plt.imshow(img, cmap='gray')
            else:
                plt.imshow(img)

        if titles and i < len(titles):
            plt.title(titles[i])

        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
## Defining a function for further use
def imshow(title='', image=None, size=10):
    W, H = image.shape[0], image.shape[1]
    aspect_ratio = H / W
    plt.figure(figsize=(aspect_ratio * size, size))
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
    
```

<h1 rtl>بهبود کیفیت و ساختار کلی یک عکس که نور کم یا مشکلاتی دارد</h1>

```python
brightness = np.ones(image.shape, dtype=np.uint8)*100
dark_pic = cv.subtract(image,brightness)
imshow("With low Bright",dark_pic)
```

```python
contrast = np.std(image)
print(contrast)
```

```python
image = image.astype(np.float32)  # تبدیل به float برای جلوگیری از اشباع

# ضرب تصویر در مقدار (مثلاً 1.5 برای افزایش کنتراست)
multiplied = image * 1.5

# کلیپ کردن مقادیر به بازه‌ی 0 تا 255 و تبدیل به uint8
contrast_img = np.clip(multiplied, 0, 255).astype(np.uint8)

cv.imshow("Original", image.astype(np.uint8))
cv.imshow("Multiplied", contrast_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
contrast = np.std(contrast_img)
print(contrast)
```

```python
enhanced = cv.convertScaleAbs(image, alpha=1.5, beta=20)

cv.imshow('Original', image)
cv.imshow('Enhanced', enhanced)
cv.waitKey(0)
cv.destroyAllWindows()
```

<h1>Cut an object from green screen</h1>

```python
Rabbit = cv.imread("Rabbit.png",-1)
lower_bond = np.array([0,200,0])
upper_bond = np.array([150,255,150])

mask = cv.inRange(Rabbit, lower_bond, upper_bond)
imshow("Rabbit", mask)
```

```python
Rabbit[mask != 0] = (0,0,0)
imshow("Rabbit", Rabbit)
```
